```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df_total = pd.read_csv('TotalData.csv').drop(['Total - 95% Lower Limit', 'Total - 95% Upper Limit'], axis=1)
df_total.head()
```

```python
age_df = pd.read_csv('AgeData.csv')
age_df = age_df.loc[10:]
age_df.head()
```

```python
educ_df = pd.read_csv('EducationData.csv')
educ_df = educ_df.loc[10:]
educ_df.head()
```

```python
gender_df = pd.read_csv('GenderData.csv')
gender_df = gender_df.loc[10:]
gender_df.head()
```

```python
race_df = pd.read_csv('RaceData.csv')
race_df = race_df.loc[10:]
race_df.head()
```

```python
dfs = [age_df, educ_df, gender_df, race_df]
mean_dfs = [df.mean().to_frame().T.loc[:, ~df.columns.str.contains('Limit')].drop('Year', axis=1) for df in dfs]
for df in mean_dfs:
    df.columns = df.columns.to_series().apply(lambda x: x.split(' -')[0])
```

```python
mean_age_df = mean_dfs[0]
mean_age_df.head()
```

```python
mean_educ_df = mean_dfs[1]
mean_educ_df.head()
```

```python
mean_gender_df = mean_dfs[2]
mean_gender_df.head()
```

```python
mean_race_df = mean_dfs[3]
mean_race_df.head()
```

```python
ax = sns.barplot(data=mean_age_df)
ax.set(xlabel='Age', ylabel='Mean Percentage Diagnosed (2010-2022)')
title = 'Mean Diabetes Rates by Age Group'
plt.title(title)
plt.savefig(title, bbox_inches='tight')
plt.show()
```

```python
ax = sns.barplot(data=mean_educ_df)
ax.set(xlabel='Education Level', ylabel='Mean Percentage Diagnosed (2010-2022)')
title = 'Mean Diabetes Rates by Education Level'
plt.title(title)
plt.savefig(title, bbox_inches='tight')
plt.show()
```

```python
ax = sns.barplot(data=mean_gender_df)
ax.set(xlabel='Gender', ylabel='Mean Percentage Diagnosed (2010-2022)')
title = 'Mean Diabetes Rates by Gender'
plt.title(title)
plt.savefig(title, bbox_inches='tight')
plt.show()
```

```python
ax = sns.barplot(data=mean_race_df)
ax.set(xlabel='Race', ylabel='Mean Percentage Diagnosed (2010-2022)')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
title = 'Mean Diabetes Rates by Race'
plt.title(title)
plt.savefig(title, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_total['Year'], df_total['Total - Percentage'], marker='o')
plt.xlabel('Year')
plt.ylabel('Diabetes Percentage')
plt.grid = True
title = 'Diabetes Over Time for US Population'
plt.title(title)
plt.savefig(title, bbox_inches='tight')
plt.show()
```

```python
education_data = pd.read_csv('EducationData.csv')[["Year", "< High School - Percentage", "High School - Percentage", "> High School - Percentage"]]
education_data.columns = ["Year", "Less than High School", "High School", "More than High School"]

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(education_data["Year"], education_data["Less than High School"], marker='o', label="Less than High School")
plt.plot(education_data["Year"], education_data["High School"], marker='o', label="High School")
plt.plot(education_data["Year"], education_data["More than High School"], marker='o', label="More than High School")

plt.title('Diabetes Rates Over Time by Education Level')
plt.xlabel('Year')
plt.ylabel('Diabetes Percentage')
plt.legend()
plt.grid(True)
plt.savefig('Diabetes Rates Over Time by Education Level', bbox_inches='tight')
plt.show()
```

```python
age_data = pd.read_csv('AgeData.csv')[['Year', '18-44 - Percentage', '45-64 - Percentage', '65-74 - Percentage', '75+ - Percentage']]

plt.figure(figsize=(12, 6))
for column in age_data.columns[1:]:
    plt.plot(age_data['Year'], age_data[column], marker='o', label=column.split()[0])

plt.title('Diabetes Rates Over Time by Age Group')
plt.xlabel('Year')
plt.ylabel('Diabetes Rate (%)')
plt.legend(title='Age Groups')
plt.grid(True)
plt.savefig('Diabetes Rates Over Time by Age Group', bbox_inches='tight')
plt.show()
```

```python
race_data = pd.read_csv('RaceData.csv')[['Year', 'Hispanic - Percentage', 'Non-Hispanic White - Percentage', 
                                'Non-Hispanic Black - Percentage', 'Non-Hispanic Asian - Percentage']]

plt.figure(figsize=(12, 6))
for column in race_data.columns[1:]:
    plt.plot(race_data['Year'], race_data[column], marker='o', label=column.split(" - ")[0])

plt.title('Diabetes Rates Over Time by Race')
plt.xlabel('Year')
plt.ylabel('Diabetes Rate (%)')
plt.legend(title='Races')
plt.grid(True)
plt.savefig('Diabetes Rates Over Time by Race', bbox_inches='tight')
plt.show()
```

```python
gender_data = pd.read_csv('GenderData.csv')[['Year', 'Male - Percentage', 'Female - Percentage']]

plt.figure(figsize=(12, 6))
for column in gender_data.columns[1:]:
    plt.plot(gender_data['Year'], gender_data[column], marker='o', label=column.split()[0])

plt.title('Diabetes Rates Over Time by Gender')
plt.xlabel('Year')
plt.ylabel('Diabetes Rate (%)')
plt.legend(title='Genders')
plt.grid(True)
plt.savefig('Diabetes Rates Over Time by Gender', bbox_inches='tight')
plt.show()
```
